```
#importing libraries
from pyspark.sql import SparkSession
!pip install yfinance
import yfinance as yf
spark = SparkSession.builder.appName('Project').getOrCreate()
```

```
#S&P 500 tickers
Ticker_df = spark.read.csv('/FileStore/tables/constituents_csv.csv',inferSchema=True,header=True)
Ticker_df.show(5)
```

```
stocks = Ticker_df.select('Symbol').collect()
```

```
# loop to extract all ticker info from yfinance and append to a list
l=[]
for i,ticker in enumerate(stocks):
  tickers = yf.Ticker(ticker.Symbol)
  try:
    keys_values = tickers.info.items()
    new_d = {str(key): str(value) for key, value in keys_values}
    l.append(new_d)
  except:
    print(ticker.Symbol)
    continue
```

```
#BRK.B -BRK-B
#BF.B -BF-B
#CTL -changed to Lumen LUM
#ETFC - merged to MS
#NBL- Moved to NASDAQ
```

```
info_df = spark.createDataFrame(l) # list to dataframe
```

```
#dataframe to CSV
info_df.repartition(1).write.csv('/FileStore/shared_uploads/saisuraja@usf.edu/info_csv_2',header=True)
```
